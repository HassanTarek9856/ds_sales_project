```python
import pandas as pd

# Load datasets
customers_df = pd.read_csv('/Users/ahmedsherif/Downloads/archive (1)/olist_customers_dataset.csv')
geolocation_df = pd.read_csv('/Users/ahmedsherif/Downloads/archive (1)/olist_geolocation_dataset.csv')
order_items_df = pd.read_csv('/Users/ahmedsherif/Downloads/archive (1)/olist_order_items_dataset.csv')
order_payments_df = pd.read_csv('/Users/ahmedsherif/Downloads/archive (1)/olist_order_payments_dataset.csv')
order_reviews_df = pd.read_csv('/Users/ahmedsherif/Downloads/archive (1)/olist_order_reviews_dataset.csv')
orders_df = pd.read_csv('/Users/ahmedsherif/Downloads/archive (1)/olist_orders_dataset.csv')
products_df = pd.read_csv('/Users/ahmedsherif/Downloads/archive (1)/olist_products_dataset.csv')
sellers_df = pd.read_csv('/Users/ahmedsherif/Downloads/archive (1)/olist_sellers_dataset.csv')
product_cat_trans_df = pd.read_csv('/Users/ahmedsherif/Downloads/archive (1)/product_category_name_translation.csv')

# 1. Merge orders with customers using 'customer_id'
df = pd.merge(orders_df, customers_df, on='customer_id', how='left')

# 2. Merge with order items using 'order_id'
df = pd.merge(df, order_items_df, on='order_id', how='left')

# 3. Merge with order payments using 'order_id'
df = pd.merge(df, order_payments_df, on='order_id', how='left')

# 4. Merge with order reviews using 'order_id'
df = pd.merge(df, order_reviews_df, on='order_id', how='left')

# 5. Merge with sellers data using 'seller_id' (from order_items)
df = pd.merge(df, sellers_df, on='seller_id', how='left')

# 6. Merge with products data using 'product_id'
df = pd.merge(df, products_df, on='product_id', how='left')

# 7. Merge with product category translation to get readable category names
df = pd.merge(df, product_cat_trans_df, on='product_category_name', how='left')

# 8. Merge with geolocation data
# Here, 'customer_zip_code_prefix' in customers_df is merged with 'geolocation_zip_code_prefix' in geolocation_df.
df = pd.merge(df,
              geolocation_df,
              left_on='customer_zip_code_prefix',
              right_on='geolocation_zip_code_prefix',
              how='left')

# Now, df is your fully merged DataFrame
df.head()
```

```python
df.drop(columns=['review_comment_message', 'review_comment_title'],inplace = True)
```

```python
for column in df.columns:
    null_count = df[column].isnull().sum()
    total_count = len(df)
    null_percentage = (null_count / total_count) * 100
    print(f"Column: {column}, Null Percentage: {null_percentage:.2f}%")
```
